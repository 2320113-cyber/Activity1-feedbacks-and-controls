```python
import serial
import time

arduino = serial.Serial('COM3', 9600)  # Open serial port
time.sleep(2)  # Wait for Arduino to reset

while True:
    command = input("Enter 1 to turn ON LED, 0 to turn OFF LED, q to quit: ")

    if command == 'q':
        break
    elif command == '1':
        arduino.write(b'1')
    elif command == '0':
        arduino.write(b'0')
    else:
        print("Invalid command. Please enter 1, 0, or q.")

arduino.close()
```
