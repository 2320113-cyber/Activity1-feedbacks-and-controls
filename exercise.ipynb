```python
import control as ctrl
import matplotlib.pyplot as plt

num =  [5, 5]
den =  [ 1, 4, 3]

G= ctrl.tf(num, den)
print(G)
```

```python
poles = ctrl.poles(G)
print ("poles:", poles)
```

```python
zeroes = ctrl.zeros(G)
print ("zeroes:", zeroes)
```

```python
ctrl.pzmap(G, plot=True)
plt.show()
```
