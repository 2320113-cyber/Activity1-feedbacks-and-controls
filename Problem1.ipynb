```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal


num = [1, 1]
den = [1, 4, 7, 6, 3]
system = signal.TransferFunction(num, den)

# Generate Step Response Data
t, y = signal.step(system)

# 1. Plot Step Response
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(t, y, 'b', linewidth=2)
plt.title('Step Response')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
# The steady state is num[-1]/den[-1] which is 1/3 (0.333)
plt.axhline(1/3, color='r', linestyle='--', label='Steady-state (0.333)')
plt.legend()

# 2. Plot Pole-Zero Map
poles = system.poles
zeros = system.zeros

plt.subplot(1, 2, 2)
plt.scatter(np.real(poles), np.imag(poles), marker='x', color='red', s=100, label='Poles')
plt.scatter(np.real(zeros), np.imag(zeros), marker='o', color='blue', s=100, facecolors='none', label='Zeros')
plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
plt.title('Pole-Zero Map')
plt.xlabel('Real Axis')
plt.ylabel('Imaginary Axis')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

# Calculate and Output Key Metrics for Verification
print(f"Poles: {poles}")
print(f"Zeros: {zeros}")
print(f"Steady-State Value: {y[-1]:.3f}")
peak_val = np.max(y)
overshoot = (peak_val - (1/3)) / (1/3) * 100
print(f"Peak Value: {peak_val:.3f}")
print(f"Percent Overshoot: {overshoot:.2f}%")
```
